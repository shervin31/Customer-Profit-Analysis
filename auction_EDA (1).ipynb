```python
# converting objects datatypes to category data types
auction['Category']=auction["Category"].astype('category')
auction['currency']=auction["currency"].astype('category')
auction['endDay']=auction["endDay"].astype('category')
```

```python
# rechecking changed data type
auction.info()
```

```python
# checking for unique category values in each categorical data

unique_categories=auction["Category"].unique()
print(unique_categories)
```

```python
for i, value in enumerate(unique_categories, 1):
            print(f"{i}. {value}")
```
